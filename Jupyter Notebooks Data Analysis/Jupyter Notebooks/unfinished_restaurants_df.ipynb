```python
import pandas as pd
```

```python
csv_path = "Resources/austin_restaurants_nutritionix.csv"

# Import the books.csv file as a DataFrame
original_df = pd.read_csv(csv_path, encoding="utf-8")
original_df.head()
```

```python
restaurants_df = original_df[["name", "brand_id", "fs_id",
                           "address", "address2", "state", "zip", "guide", "id", "lat",
                           "lng"]]


restaurants_df = restaurants_df.rename(columns={"name": "Name",
                                               "brand_id": "Brand Id", 
                                               "fs_id": "Fs Id",
                                               "address": "Address", 
                                               "address2": "Address 2", 
                                               "state": "ST", 
                                               "zip": "Zip", 
                                               "guide": "Guide", 
                                               "id": "Id", 
                                               "lat": "Lat",
                                               "lng": "Lng"})
#Sorted by Name of Restaurant
restaurants_df.sort_values("Name", inplace = True) 
  
# dropping ALL duplicte values 
restaurants_df.drop_duplicates(subset ="Name", inplace = True) 
restaurants_df

```
