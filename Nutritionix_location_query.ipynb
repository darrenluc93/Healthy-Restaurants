```python
import requests
import pandas as pd
from config import app_id, api_key
import numpy as np
```

```python
base_url = 'https://trackapi.nutritionix.com/'
headers = {'x-app-id': app_id, 'x-app-key':api_key}
```

```python
full_url = f"{base_url}v2/locations"
full_url
```

```python
#north_bound, south_bound, east_bound, west_bound = 30.5, 30.0, -97.4, -98.1
locations_list = []
for i in np.linspace(30.0, 30.5, 5):
    for j in np.linspace(-97.4, -98.1, 5):
        params = {'ll': f"{i}, {j}", 'limit': 100, 'distance': '10mi'}
        response = requests.get(f"{base_url}v2/locations", headers=headers, params=params).json()
        locations_list.append(response)
```

```python
locations_list[0]
```

```python
full_locations = [locations_list[i]['locations'][j] for i in range(len(locations_list)) for j in range(len(locations_list[i]['locations']))]
```

```python
full_locations[0]
```

```python
full_locations_df = pd.DataFrame(full_locations)
```

```python
full_locations_df.to_csv('full_locations_greater_austin_area.csv')
```

```python
austin_from_full_df = full_locations_df[full_locations_df['city'] == 'Austin']
```
